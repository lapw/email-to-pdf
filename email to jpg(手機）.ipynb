```python
text='''@題目：日記一則

以下列文字作文章開首：

今天這事情的結果是我始料不及的，如果我早一些明白「三省其身」的道理，結果該不至於這樣。@

@1）《論語·學而》：「曾子曰：『吾日三省吾身，為人謀而不忠乎？與朋友交而不信乎？傳不習乎？』」 省：檢查、反省；身：自身。 原指每日從三個方面檢討自己，後指多次自覺地自我反省。@

@2）選材

一定要是今天內發生的，原因可以是之前的，但經過、結果最好在今天發生。可以是一場比賽、一次測驗、考試面試等等。@

@3）倒敍

倒敍法就是在文章的開頭先寫事情的結果，再寫事情的原因、經過。這樣的描述方式有別於「順敍」（先發生的先寫，後發生的後寫）的方式。把事件倒過來寫，把結果放在最前面，有懸疑的效果，引發讀者的好奇心，讓人想一窺究竟。@

@4）插敍

插敍是在敍述事件的經過中，為了幫助展開情節而暫時中斷敍述，插入一段與主要情節相關的回憶或故事，對主要情節起補充、襯托的作用。在插敍的前面和後面，要用適當的文字過渡，令文章更流暢自然。@

@5）加入人物的行為、動作、說話，令情節更生動，例如加入被師長責備、同學取笑，或師長、同學安慰、鼓勵的話，令情節更豐富。

可用詞語：人誰無過、再接再厲、重整旗鼓、不屈不撓、知恥近乎勇@

@6）感受、反省

說出這次事件帶給自己的感受，是大失所望？後悔莫及？悔不當初？反省自己為什麼會這樣做、做錯了甚麽、以後應該怎樣做。@'''
```

```python
s1 = re.findall (r"@([\S\s]*?)@",text)

txBox = {}
tf = {}


print(s1)
```

```python
  prs = Presentation()
  r = glob.glob("*.jpg")
  sr =random.choice(r)
```

```python
#slide張數
i = len(s1)
print(len(s1))
```

```python
 #制slide
for num in range(i):
    
  title_slide_layout = prs.slide_layouts[6]
  slide = prs.slides.add_slide(title_slide_layout)

  left = top = Inches(0)
  img_path = str(sr)
  pic = slide.shapes.add_picture(img_path, left, top, width=prs.slide_width, height=prs.slide_height)
  
  # This moves it to the background
  slide.shapes._spTree.remove(pic._element)
  slide.shapes._spTree.insert(2, pic._element)
```

```python
###第一個文字方塊
left = top = Inches(1) #設定文字方塊擺放位置，文字方塊距離當前投影片左方及上方的距離
width = Inches(8) 
height = Inches(5) #設定文字方塊大小
txBox[0] = prs.slides[0].shapes.add_textbox(left, top, width, height) #依照設定加入空白文字方塊
tf[0] = txBox[0].text_frame #取文字方塊的文字框物件
tf[0].paragraphs[0].text = s1[0] #在文字框物件輸入文字，此段文字為文字方框的第一段落
tf[0].paragraphs[0].font.size = Pt(50)    # 大小
tf[0].vertical_anchor = MSO_ANCHOR.MIDDLE
tf[0].paragraphs[0].font.name = '微软雅黑'
```

```python
##第二張之後文字方塊
for sli in range(1,i):
  left =  Inches(1) #設定文字方塊擺放位置，文字方塊距離當前投影片左方及上方的距離
  top = Inches(0.5)
  width = Inches(8) 
  height = Inches(6) #設定文字方塊大小
  txBox[sli] = prs.slides[sli].shapes.add_textbox(left, top, width, height) #依照設定加入空白文字方塊
  tf[sli] = txBox[sli].text_frame #取文字方塊的文字框物件
  tf[sli].paragraphs[0].text = s1[sli] #在文字框物件輸入文字，此段文字為文字方框的第一段落
  tf[sli].paragraphs[0].font.size = Pt(35)    # 大小
  tf[sli].vertical_anchor = MSO_ANCHOR.MIDDLE

  tf[sli].paragraphs[0].font.name = '微软雅黑'
```

```python
#Final step = Creation and saving of pptx
prs.save('test.pptx')
```
